```python
import pandas as pd
from tqdm import tqdm
```

```python
ultrafeedback = pd.read_csv("kaggle-ultrafeedback-drop-duplicate.csv")
ultrafeedback.head()
```

```python
ultrafeedback_tie = pd.read_csv("kaggle-ultrafeedback-ties-drop-duplicate.csv")
ultrafeedback_tie.head()
```

```python
ultrafeedback_combined = pd.concat([ultrafeedback, ultrafeedback_tie], ignore_index=True)
ultrafeedback_combined.tail()
```

```python
original_train = pd.read_csv("../data/train.csv")
original_train.head()
```

```python
prompt_sets = set()
duplicate_prompt_counts = 0

duplicate_index = []

for idx1, row1 in tqdm(original_train.iterrows(), total=len(original_train)):
    if row1["prompt"] in prompt_sets:
        # print(idx1)
        continue
    else:
        prompt_sets.add(row1["prompt"])

for idx2, row2 in tqdm(ultrafeedback_combined.iterrows(), total=len(ultrafeedback_combined)):
    if row2["prompt"] in prompt_sets:
        duplicate_prompt_counts += 1
        duplicate_index.append(idx2)
        # print(idx2, row2["prompt"])

print(duplicate_prompt_counts)

ultrafeedback_combined.drop(duplicate_index, inplace=True)
ultrafeedback_combined.reset_index(drop=True, inplace=True)
print(ultrafeedback_combined.shape)
```
