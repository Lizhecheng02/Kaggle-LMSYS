```python
from tqdm import tqdm
import pandas as pd
import re
import json
```

```python
extra = pd.read_csv("lmsys-chatbot_arena_conversations-33k.csv")
extra.head()
```

```python
def unicode_preprocess(input_str):
    return json.loads(input_str)

extra.loc[:, "prompt"] = extra["prompt"].apply(unicode_preprocess)
extra.loc[:, "response_a"] = extra["response_a"].apply(unicode_preprocess)
extra.loc[:, "response_b"] = extra["response_b"].apply(unicode_preprocess)
```

```python
def contains_unicode_pattern(s):
    pattern = r"\\u[0-9a-fA-F]{4}"
    match = re.search(pattern, s)
    return match is not None
```

```python
s1 = r"This string contains \u043c\u043b"
s2 = r"This string does not contain the pattern"

print(contains_unicode_pattern(s1))
print(contains_unicode_pattern(s2))
```

```python
problem_idx_sets = set()

for idx, row in tqdm(extra.iterrows(), total=len(extra)):
    for i in range(len(row["prompt"])):
        if contains_unicode_pattern(row["prompt"][i]):
            print(row["prompt"][i], idx)
            problem_idx_sets.add(idx)
    for i in range(len(row["response_a"])):
        if contains_unicode_pattern(row["response_a"][i]):
            print(row["response_a"][i], idx)
            problem_idx_sets.add(idx)
    for i in range(len(row["response_b"])):
        if contains_unicode_pattern(row["response_b"][i]):
            print(row["response_b"][i], idx)
            problem_idx_sets.add(idx)

print(len(problem_idx_sets))
print(problem_idx_sets)
```

```python
extra = pd.read_csv("lmsys-chatbot_arena_conversations-33k.csv")

delete_idx_list = sorted(problem_idx_sets)
print(delete_idx_list)
extra.drop(delete_idx_list, inplace=True)
extra.reset_index(drop=True, inplace=True)
extra.to_csv("lmsys-chatbot_arena_conversations-33k_clean.csv", index=False)
print(extra.shape)
```
